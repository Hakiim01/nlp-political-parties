# Topic Extraction
Convert the corpus into a format suitable for topic modeling. Each document should be a single string, and we will group by `Text_ID` to combine all text entries for each document.

```python
from tqdm import tqdm
import json

tqdm.pandas()

grouped = (
    corpus_df
    .groupby("Text_ID")
    .progress_apply(lambda g: " ".join(g["text"]))
    .reset_index()
    .rename(columns={0: "text"})
)
```
